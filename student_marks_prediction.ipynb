## **We can  see that variables are linearlly related to each other so we can use LInear regression method to build this model**

#### Linear Regression  - 
In statistics, linear regression is a linear approach to modelling the relationship between a scalar response and one or more explanatory variables (also known as dependent and independent variables).

In simple we will predict best fit line `y=mx+c` for our data.
Here
*   y = marks [target value] 
*   x = hours[input value]
*   c = y-intercept [actual - prediction]
*   m = weight[slope of predicted line]

We will find these weight and bias values and based on that we will predict the result.
linear_reg.png

```python
#lets import LinearRegression class from scikit learn
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(raw_df[['Hours']],raw_df['Scores'])
```

```python
inputs = raw_df[['Hours']]
```

```python
predictions = model.predict(inputs)
```

```python
raw_df['Predicted']=pd.DataFrame((np.ceil(predictions)).astype(int))
```

```python
raw_df
```

# **Task : Predict Score If Student does study for 9.25 hrs/day ?**

```python
x= model.predict([[9.25]])
```

```python
x
```

##  **So our Model Predicted that student can get around 92.90(93) marks if studies for 9.25 hrs/daily.** 

### Predicting Marks With INPUTS

```python
def predict_marks(hrs):
    mod = model.predict([[hrs]])
    mod = np.ceil(mod).astype(int)
    if(mod>100):
        mod = 100
    print("You can get approx {} marks if you study {} hrs daily.".format(mod,hrs))
```

```python
predict_marks(12.25)
```

```python
inp = float(input("Enter Hours Of Study "))
predict_marks(inp)
```

## **Lets see how our predicted line fits the data**


```python
plt.figure(figsize=(12,6))

plt.scatter(raw_df['Hours'],raw_df['Scores'])
plt.plot(raw_df['Hours'],raw_df['Predicted'],color='red')
plt.legend(['Predicted','Actual']);
```
